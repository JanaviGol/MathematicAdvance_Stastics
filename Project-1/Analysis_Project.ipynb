## Analysis Income Distribution and Household Demographics

### 1. Library Imports and Data Loading

```python
# Importing essential libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis
```

```python
# loading the dataset
df = pd.read_csv('household_demographics_data.csv')
```

### 2. Types of Data Category

```python
numerical_cols = df.select_dtypes(include=['number']).columns.tolist()

categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

print("-" * 30)
print(f" Numerical Columns: {numerical_cols}")
print(f" Categorical Columns: {categorical_cols}")
print("-" * 30)
```

### 3. Central Tendency Analysis

```python
# Calculating Mean, Median, and Mode for Income 
income_mean = df['Household_Income'].mean()
income_median = df['Household_Income'].median()
income_mode = df['Household_Income'].mode()[0]  

print("--- INCOME RESULTS ---")
print("Average Income (Mean):", income_mean)
print("Middle Income (Median):", income_median)
print("Most Common Income (Mode):", income_mode)
print("----------------------")
```

```python
# Calculating Mean, Median, and Mode for Age
age_mean = df['Age_of_Household_Head'].mean()
age_median = df['Age_of_Household_Head'].median()
age_mode = df['Age_of_Household_Head'].mode()[0]

print("--- AGE RESULTS ---")
print("Average Age (Mean):", age_mean)
print("Middle Age (Median):", age_median)
print("Most Common Age (Mode):", age_mode)
print("-------------------")
```

### 4. Measures of Dispersion

```python
# Calculating Range, Variance, and Standard Deviation
inc_min = df['Household_Income'].min()
inc_max = df['Household_Income'].max()

income_range = inc_max - inc_min
income_variance = df['Household_Income'].var()
income_std_dev = df['Household_Income'].std()

print("--- Income Spread Results ---")
print("1. Min Income:", inc_min)
print("2. Max Income:", inc_max)
print("3. Range:", income_range)
print("4. Variance:", income_variance)
print("5. Standard Deviation:", income_std_dev)
```

```python
# Interquartile Range (IQR) 
Q1 = df['Household_Income'].quantile(0.25)
Q3 = df['Household_Income'].quantile(0.75)

IQR = Q3 - Q1

print("--- IQR RESULTS ---")
print("1. Q1 (25th Percentile):", Q1)
print("3. Q3 (75th Percentile):", Q3)
print("4. Interquartile Range (IQR):", IQR)
print("-------------------")
```

```python
print("COMMMENT ON INCOME SPREAD:")
print(f"The middle 50% of households earn between {Q1} and {Q3}.")
print(f"The IQR value is {IQR}. This tells us how much the middle-class income varies.")
```

### 5: Skewness and Kurtosis

```python
# Histogram : Income Distribution with Gaussian Curve
plt.figure(figsize=(10, 6))
sns.histplot(df['Household_Income'], kde=True, color='skyblue', stat="density")
plt.title('Income Distribution with Gaussian Curve')
plt.xlabel('Household Income')
plt.ylabel('Density')
plt.show()

# Checking the shape of the income distribution
income_skew = df['Household_Income'].skew()
income_kurt = df['Household_Income'].kurt()

print("--- DISTRIBUTION STATS ---")
print(f"1. Skewness: {income_skew:.2f}")
print(f"2. Kurtosis: {income_kurt:.2f}")
print("--------------------------")
```

### 6.Visualizations 

```python
# Box plots : Income Comparison by Edecation Level
plt.figure(figsize=(10, 6))
sns.boxplot(x='Education_Level', y='Household_Income', data=df, palette='Set2')
plt.title('Income Comparison by Education Level')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Layered Histogram with KDE (Kernel Density Estimation) : Income Distribution across Education Levels
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Household_Income', hue='Education_Level', kde=True, element="step")
plt.title('Income Distribution across Education Levels')
plt.show()
```

```python
# Histogram & KDE : Distribution of Family Size
plt.figure(figsize=(8, 5))
sns.histplot(df['Family_Size'], kde=True, color='green')
plt.title('Distribution of Family Size (Histogram & KDE)')
plt.show()
```

```python
# Box plots : Family Size Conparison by Education Level 
plt.figure(figsize=(10, 6))
sns.boxplot(x='Education_Level', y='Family_Size', data=df, palette='Pastel1')
plt.title('Family Size Comparison by Education Level')
plt.show()
```

```python
# Scatter Plot with Regression Line : Age vs Household Income Distribution 
plt.figure(figsize=(10, 6))
sns.regplot(x='Age_of_Household_Head', y='Household_Income', data=df, 
            scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Age vs Household Income Distribution')
plt.show()
```
