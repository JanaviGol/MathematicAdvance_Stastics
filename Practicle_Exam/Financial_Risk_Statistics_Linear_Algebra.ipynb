## 📌 Section 1: Business Statistics

### 1. Central Tendency
Central Tendency represents the center point or a single typical value of a data distribution. It helps summarize an entire dataset using one representative value.

* **🧮 Mean:** The mathematical average calculated by dividing the sum of all values by the total number of data points.
  * *Example:* If we add Rahul's, Rohit's, and a third applicant's income (\$75,000) and divide by 3, the average bank customer income is $\text{Mean} = \$60,000$.
* **📍 Median:** The exact middle value when data points are arranged in ascending or descending order. It is highly robust against extreme outliers.
  * *Example:* If a billionaire applies for a loan at this branch, the **Mean** income shoots up unnaturally, but the **Median** remains right in the middle, representing the true common man.
* **🔄 Mode:** The value or category that appears most frequently within the dataset.
  * *Example:* If most applicants in our dataset request a loan amount of exactly \$15,000, then \$15,000 is the **Mode**.

---

### 2. Dispersion
Dispersion measures how spread out or scattered the data points are around the center point. It shows the level of variability within the dataset.

* **↔️ Range:** The simplest measure of spread, calculated as the absolute difference between the maximum and minimum values in the dataset.
  * *Example:* If the highest loan request is \$200,000 and the lowest is \$5,000, the **Range** of borrowing is \$195,000.
* **📉 Variance:** The average of the squared differences from the Mean. It measures the overall volatility of the data points.
* **📊 Standard Deviation:** The square root of the variance. It expresses the spread in the exact same units as the original data, making it highly intuitive for interpretation.
  * *Example:* A low standard deviation in customer credit scores means most applicants have very similar credit behavior; a high standard deviation means the risk varies drastically from customer to customer.

---

### 3. Skewness and Kurtosis
These metrics describe the structural shape, symmetry, and tail-behavior of a data distribution curve.

* **📐 Skewness:** Measures the lack of symmetry in a distribution.
  * ➡️ **Positive Skew:** Tail extends to the right side (higher values). *Example: Most customers have low-to-medium incomes, while a few ultra-rich people pull the tail to the right.*
  * ⬅️ **Negative Skew:** Tail extends to the left side (lower values).
  * 🟢 **Zero Skew:** Perfectly symmetrical bell curve.
* **🔔 Kurtosis:** Measures the heaviness of the tails or the peakedness of the distribution curve relative to a normal distribution.
  * ⛰️ **Leptokurtic (High):** Heavy tails with a sharp, high peak. *(Risk is concentrated in specific extreme groups).*
  * 🕳️ **Platykurtic (Low):** Light tails with a flatter, broader peak. *(Risk is evenly distributed).*
  * 🪵 **Mesokurtic:** Standard normal distribution shape.

---

### 4. Probability Concepts
Probability provides a mathematical framework to evaluate risk and uncertainty in business decision-making.

* **🎯 Baseline Probability:** The independent, general likelihood of an event occurring over the entire population without considering any external conditions.
  * *Example:* The default rate across the entire bank history is 10%. So, any random user has a baseline default probability of $P(\text{Default}) = 0.10$.
* **🔍 Conditional Probability:** The likelihood of an event occurring given that another specific condition or event has already taken place ($P(A \mid B)$).
  * *Example:* What is the probability that a customer defaults *given that* their credit score is less than 600? This isolates risk into high-probability segments: $P(\text{Default} \mid \text{Credit Score} < 600)$.
* **📋 Contingency Table (Crosstab):** A matrix layout or table format used to analyze the relationship between multiple categorical variables by displaying their joint frequencies.
  * *Example:* A grid showing exactly how many 'Low Credit Score' users said 'Yes' vs 'No' to a loan default.

## 📌 Section 2: Linear Algebra

### 1. Vector Space and Geometry
Linear Algebra allows us to treat complex multi-attribute customer profiles as individual geometric points in a high-dimensional vector space.

* **🚀 Vector Representation:** Combining distinct customer attributes to form a structured multi-dimensional numerical vector.
  * *Example:* We can map Rahul and Rohit into 2D financial space where the X-axis is `Income` and the Y-axis is `Loan_Amount`:
    $$\vec{v}_{\text{Rahul}} = \begin{bmatrix} 60000 \\ 20000 \end{bmatrix}, \quad \vec{v}_{\text{Rohit}} = \begin{bmatrix} 45000 \\ 15000 \end{bmatrix}$$
* **📏 L2 Norm (Euclidean Norm):** Measures the absolute magnitude, length, or scale of a vector from the origin, calculated as the square root of the sum of squared vector components.
  * *Example:* The L2 Norm represents the gross financial scale of an applicant. Rahul's absolute financial weight is calculated as:
    $$\|\vec{v}_{\text{Rahul}}\| = \sqrt{60000^2 + 20000^2}$$

---

### 2. Angular and Matrix Relationships
These concepts help calculate similarity and process transformations across large financial data matrices.

* **✖️ Dot Product:** The sum of the products of corresponding components across two vectors. It serves as the foundation for calculating vector projections and similarity.
  * *Example:* Multiplying corresponding financial attributes of Rahul and Rohit together:
    $$\vec{v}_{\text{Rahul}} \cdot \vec{v}_{\text{Rohit}} = (60000 \times 45000) + (20000 \times 15000)$$
* **📐 Cosine Theta & Directional Angle:** Evaluates the geometric orientation between two customer profiles.
  * *Example:* Even though Rahul makes more money than Rohit, **both ask for a loan that is exactly $1/3$ of their income**. Because their financial proportions are identical, their vectors point in the exact same direction, resulting in a **Directional Angle of $0^\circ$ (Cosine Theta = 1)**. This shows they have identical financial behavior profiles!
